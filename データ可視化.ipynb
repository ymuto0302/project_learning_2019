# 折れ線グラフ

```
import matplotlib.pyplot as plt
x = [4, 3, 4, 6, 7]
y = [2, 0, 2, 3, 2]
plt.plot(x)
plt.plot(y)
plt.show()
```

# 放物線
\begin{align}
y=x^2 \quad (-3 \le x \le 3)
\end{align}

```
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(-3, 3, 0.1)
y = x**2
plt.plot(x, y)
plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt
data = np.random.randn(10000) # 1万件の正規乱数
plt.hist(data, bins=50)
plt.show()
```
